````python
import pandas as pd
import matplotlib.pyplot as plt

# Paso 1: Crear un diccionario con países europeos, poblaciones por género y edad
paises_poblacion = {
    "Alemania": {
        "Hombres": {"0-18": 8.0, "19-64": 35.0, "65-150": 9.51},
        "Mujeres": {"0-18": 7.5, "19-64": 33.0, "65-150": 10.01}
    },
    "Francia": {
        "Hombres": {"0-18": 7.2, "19-64": 31.5, "65-150": 8.2},
        "Mujeres": {"0-18": 6.8, "19-64": 29.9, "65-150": 9.49}
    },
    "Italia": {
        "Hombres": {"0-18": 6.0, "19-64": 29.2, "65-150": 8.5},
        "Mujeres": {"0-18": 5.7, "19-64": 28.0, "65-150": 10.1}
    },
    "España": {
        "Hombres": {"0-18": 4.9, "19-64": 23.5, "65-150": 7.0},
        "Mujeres": {"0-18": 4.6, "19-64": 22.3, "65-150": 7.05}
    },
    "Países Bajos": {
        "Hombres": {"0-18": 1.9, "19-64": 8.3, "65-150": 1.5},
        "Mujeres": {"0-18": 1.8, "19-64": 7.9, "65-150": 1.4}
    }
}

# Paso 2: Convertir el diccionario a un DataFrame de pandas para facilitar el manejo de datos
poblacion_df = pd.DataFrame({
    (pais, genero): data
    for pais, generos in paises_poblacion.items()
    for genero, data in generos.items()
}).T  # Transponer para mejor visualización

# Mostrar el DataFrame
print(poblacion_df)

# Paso 3: Visualización de la población total por país (sumando hombres y mujeres)

# Sumar las poblaciones por grupo de edad y género
poblacion_total = poblacion_df.sum(axis=1).unstack(level=1).sum(axis=1)

# Crear gráfico de barras para la población total por país
poblacion_total.plot(kind='bar', color='lightgreen')

# Añadir título y etiquetas
plt.title("Población Total de Países Europeos (en millones)")
plt.xlabel("País")
plt.ylabel("Población Total (millones)")

# Mostrar el gráfico
plt.tight_layout()
plt.show()

# Paso 4: Visualización de la población por género y grupo de edad para un país específico (Alemania)

# Seleccionar los datos de Alemania
poblacion_alemania = poblacion_df.loc["Alemania"]

# Crear un gráfico de barras apilado para ver hombres y mujeres en cada grupo de edad
poblacion_alemania.plot(kind='bar', stacked=True)

# Añadir título y etiquetas
plt.title("Población de Alemania por género y grupo de edad")
plt.xlabel("Grupo de Edad")
plt.ylabel("Población (millones)")

# Mostrar el gráfico
plt.tight_layout()
plt.show()


````
